```python
from sklearn.feature_extraction import DictVectorizer
```

```python
import numpy as np
categorical_features = X.select_dtypes(exclude=[np.number])
print(categorical_features)
features_dict = categorical_features.to_dict(orient='records')
transformer = DictVectorizer(sparse=False)
categorical_features_encoded = transformer.fit_transform(features_dict)
categorical_features_encoded
```

```python
numeric_features = X.select_dtypes(include=[np.number])
numeric_features
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
numeric_features_scaled = scaler.fit_transform(numeric_features)
```

```python
X_final = np.concatenate((numeric_features_scaled, categorical_features_encoded), axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y, test_size=0.3, random_state=42)
```

```python
from sklearn.svm import SVC
clf = SVC(probability=True)
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score, precision_recall_curve, auc
import matplotlib.pyplot as plt
y_pred = clf.predict_proba(X_test)[:, 1]
fpr, tpr, _ = roc_curve(y_test, y_pred)
precision, recall, _ = precision_recall_curve(y_test, y_pred)

roc_auc = roc_auc_score(y_test, y_pred)
pr_auc = auc(recall, precision)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title(f'ROC Curve (AUC = {roc_auc:.2f})')

#====================#

plt.subplot(1, 2, 2)
plt.plot(recall, precision)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title(f'Precision-Recall Curve (AUC = {pr_auc:.2f})')

plt.tight_layout()
plt.show()
```

```python
best_clf = SVC(**best_params, probability=True)
best_clf.fit(X_train, y_train)

y_pred_best = best_clf.predict_proba(X_test)[:, 1]
roc_auc_best = roc_auc_score(y_test, y_pred_best)

print(f'Ð›ÑƒÑ‡ÑˆÐ¸Ðµ Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ñ Ð¿Ð°Ñ€Ð°Ð¼ÐµÑ‚Ñ€Ð¾Ð²: {best_params}')
print(f'AUC-ROC Ð½Ð° Ñ‚ÐµÑÑ‚Ð¾Ð²Ð¾Ð¹ Ð²Ñ‹Ð±Ð¾Ñ€ÐºÐµ Ñ Ð»ÑƒÑ‡ÑˆÐ¸Ð¼Ð¸ Ð¿Ð°Ñ€Ð°Ð¼ÐµÑ‚Ñ€Ð°Ð¼Ð¸: {roc_auc_best:.2f}')
```
